```python
### LOOKING AT A WEBSITE WITH REQUESTS

# We can start by looking at a website. 
# The object, r, will contain the website at the time when we scraped it.

import requests
r = requests.get('https://boston.craigslist.org/search/aap') #Craiglist boston apartments

#We can see a few things from this: 
# 1. The status code:

print(r.status_code)
```

```python
# 2. See if that status code is alright

r.status_code == requests.codes.ok

# If this outputs, true, it means the webpage is available. 
```

```python
# 3. Look at the text

r.text
```

```python
# But that is hard to read! We need to parse it out.

from bs4 import BeautifulSoup as bs4

html = bs4(r.text, 'html.parser')
print(html.prettify()[:1000])

#Much nicer to read!
```

```python
#We can also find the apartments in this cleaned bit of of text

price_apts = html.find_all(attrs={'class':'price'})

```

```python
## There are libraries that can help with this, beyond just beautiful soup and requests. 

!pip install python-craigslist
from craigslist import CraigslistJobs, CraigslistForSale, CraigslistHousing
CraigslistJobs.show_filters()
```

```python
## We can look at apartments much easier now

cl_h = CraigslistHousing(site='sfbay', area='sfc', category='roo',
                         filters={'max_price': 1200, 'private_room': True})
for result in cl_h.get_results(sort_by='newest', geotagged=True):
    print(result)
```
